3) Criar um scipt para realizar as normalizações: MinMax, Valor Máximo e por Desvio Padrão dos dados abaixo. Mostre o antes e o depois.

```python
base_path = ("/content/drive/My Drive/inteligencia_artificial/ia_cpf.csv")
df_cpf = pd.read_csv(base_path)
```

```python
df_cpf.head()
```

```python
#MinMax
norm_min_max = df_cpf.copy()
for column in norm_min_max.columns:
    norm_min_max[column] = (norm_min_max[column] - norm_min_max[column].min()) / (norm_min_max[column].max() - norm_min_max[column].min())
print(norm_min_max)
```

```python
#Valor máximo
norm_max = df_cpf.copy()
for column in norm_max.columns:
    norm_max[column] = norm_max[column]  / norm_max[column].max()

display(norm_max)
```

```python
#Desvio Padrão
print("Média:", np.mean(df_cpf.despesa))

print("Desvio padrão:", np.std(df_cpf.despesa))
```
