```python
#from sklearn.datasets import load_abalone
import pandas as pd
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data'
attributes = ["Sex", "Length", "Diameter_height", "Whole_weight", "Shucked_weight", "Viscera_weight", "Shell_weight", "Rings"]
abalone = pd.read_csv(url, names = attributes)
```

```python
print(abalone.count()) #tamanho da base de dados e variaveis existentes
```
